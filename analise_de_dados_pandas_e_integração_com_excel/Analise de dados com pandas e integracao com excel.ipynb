# Comparando, Tratando e Mesclando DataFrames
## Objetivo
Vamos modificar os IDs para os nomes dos produtos, dos clientes e das lojas, para nossas análises ficarem mais intuitivas futuramente. Para isso, vamos criar uma data frame com todos os detalhes.
- Usaremos o método merge para isso e, depois se quisermos, podemos pegar apenas as colunas que queremos do dataframe final.

## Criando nossos dataframes

````python
import pandas as pd

vendas_df = pd.read_csv(r'arquivos/Contoso - Vendas - 2017.csv', sep=';')
produtos_df = pd.read_csv(r'arquivos/Contoso - Cadastro Produtos.csv', sep=';')
lojas_df = pd.read_csv(r'arquivos/Contoso - Lojas.csv', sep=';')
clientes_df = pd.read_csv(r'arquivos/Contoso - Clientes.csv', sep=';')

display(vendas_df)
display(produtos_df)
display(lojas_df)
display(clientes_df)
````

## Vamos tirar as colunas inúteis do clientes_df ou pegar apenas as colunas que quisermos

```
.drop([coluna1, coluna2, coluna3]) -> retira as colunas: coluna1, coluna2 e coluna3
```

````python
# axis 1 = coluna
# axis 0 = linha
# retirar colunas
clientes_df = clientes_df.drop(['Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9', 'Unnamed: 10'], axis=1)
display(clientes_df)
````

````python
# deixar somente as colunas
clientes_df = clientes_df[['ID Cliente', 'E-mail']]
produtos_df = produtos_df[['ID Produto', 'Nome da Marca']]
lojas_df = lojas_df[['ID Loja', 'Nome da Loja']]
display(produtos_df)
````

## Agora vamos juntar dataframes para ter um unico dataframe com tudo

``
novo_dataframe = dataframe1.merge(dataframe2, on='coluna')
``

- Obs: O merge precisa das colunas com o mesmo nome para funcionar. Se não tiver, você precisa alterar o nome da coluna com .rename

``
dataframe.rename(columns={'coluna1': 'nova_coluna1'})
``

````python
# juntando dataframes
vendas_df = vendas_df.merge(produtos_df, on='ID Produto')
vendas_df = vendas_df.merge(lojas_df, on='ID Loja')
vendas_df = vendas_df.merge(clientes_df, on='ID Cliente')


display(vendas_df)
````

````python
vendas_df = vendas_df.rename(columns={'E-mail': 'E-mail Cliente'})
display(vendas_df)
````

# Resumos e um pouco de Visualização no pandas
## Resumo
Vamos ver alguns metodos para analisar nossas tabelas(dataframes)<br>
Além disso, vamos usar os plots de gráficos padrões do pandas.<br>
OBS: O pandas usa o matplotlib para plotar graficos.<br>
- Preparando as bases de dados


````python
import pandas as pd
# importando arquivos
vendas_df = pd.read_csv(r'arquivos/Contoso - Vendas - 2017.csv', sep=';')
produtos_df = pd.read_csv(r'arquivos/Contoso - Cadastro Produtos.csv', sep=';')
lojas_df = pd.read_csv(r'arquivos/Contoso - Lojas.csv', sep=';')
clientes_df = pd.read_csv(r'arquivos/Contoso - Clientes.csv', sep=';')

# limpando apenas as colunas que queremos
clientes_df = clientes_df[['ID Cliente', 'E-mail']]
produtos_df = produtos_df[['ID Produto', 'Nome da Marca']]
lojas_df = lojas_df[['ID Loja', 'Nome da Loja']]

# mesclando e renomeando os dataframes
vendas_df = vendas_df.merge(produtos_df, on='ID Produto')
vendas_df = vendas_df.merge(lojas_df, on='ID Loja')
vendas_df = vendas_df.merge(clientes_df, on='ID Cliente')
vendas_df = vendas_df.rename(columns={'E-mail': 'E-mail Cliente'})
display(vendas_df)
````

## Qual cliente que comprou  mais vezes?
- Usaremos o metodo .value_counts() para contar quantas vezes cada valor do Datadrame aparece
- Usaremos o metodo .plot() para exibir um grafico.

````python
frequencia_clientes = vendas_df['E-mail Cliente'].value_counts()
display(frequencia_clientes)
frequencia_clientes.head().plot(figsize=(15, 5), yticks=range(65, 80, 2))
````

## Qual loja vendeu mais?
- Usaremos o .groupby para agrupar nosso dataframe, de acordo com o que queremos(somando a quantidade de vendas por exemplo)

````python
vendas_loja = vendas_df.groupby('Nome da Loja').sum()
vendas_loja = vendas_loja[['Quantidade Vendida']]
display(vendas_loja)
````

- Agora precisamos pegar o maior valor. Temos 2 formas
    1. Ordenar o dataframe em ordem decrescente de Quantidade Vendidas
        - metodo .sort_values
    2. Pegar o Maior valor diretamente
        - metodo .max() e idxmax()

````python
# ordenando o dataframe
vendas_loja = vendas_loja.sort_values('Quantidade Vendida', ascending=False)
display(vendas_loja)
# podemos plotar em um grafico
vendas_loja.head().plot(figsize=(15, 5), kind='bar')
````

````python
# pegando maior valor e de indice
maior_valor = vendas_loja['Quantidade Vendida'].max()
melhor_loja = vendas_loja['Quantidade Vendida'].idxmax()
print(maior_valor, melhor_loja)
````

## QUal produto que menos vendeu?
- Já temos uma lista criada para isso, basta verificarmos o final da lista(que já esta ordenada) ou usamos os metodos:
    1. min()
    2. idxmin()

````python
menor_valor = vendas_loja['Quantidade Vendida'].min()
pior_loja = vendas_loja['Quantidade Vendida'].idxmin()
print(menor_valor, pior_loja)
````

# Filtrando informações nos dataframes
## Resumo
Um dos grandes potenciais do pandas é para tratar condições<br>
E a forma com que analisamos condições no dataframe é diferente do que já fizemos até agora.
- Preparando base de dados

````python
import pandas as pd
# importando arquivos
vendas_df = pd.read_csv(r'arquivos/Contoso - Vendas - 2017.csv', sep=';')
produtos_df = pd.read_csv(r'arquivos/Contoso - Cadastro Produtos.csv', sep=';')
lojas_df = pd.read_csv(r'arquivos/Contoso - Lojas.csv', sep=';')
clientes_df = pd.read_csv(r'arquivos/Contoso - Clientes.csv', sep=';')

# limpando apenas as colunas que queremos
clientes_df = clientes_df[['ID Cliente', 'E-mail']]
produtos_df = produtos_df[['ID Produto', 'Nome da Marca']]
lojas_df = lojas_df[['ID Loja', 'Nome da Loja']]

# mesclando e renomeando os dataframes
vendas_df = vendas_df.merge(produtos_df, on='ID Produto')
vendas_df = vendas_df.merge(lojas_df, on='ID Loja')
vendas_df = vendas_df.merge(clientes_df, on='ID Cliente')
vendas_df = vendas_df.rename(columns={'E-mail': 'E-mail Cliente'})
display(vendas_df)
````

## Primeiro, vamos aplicr uma função normalmente. Qual o % das vendas foram devolvidas?
- Para isso vamos somar as quantidades nas colunas correpondentes. Lembrando, o % vai ser: Total Devolvido/ Total Vendido

````python
qtde_vendida = vendas_df['Quantidade Vendida'].sum()
qtde_devolvida = vendas_df['Quantidade Devolvida'].sum()

print(f'{qtde_devolvida/qtde_vendida:.2%}')
````

## Agora, se quisermos fazer a mesma análise apenas para uma loja. Queremos filtrat apenas os itens da Loja Contoso Europe Oline e saber o % de devolução dessa loja.
- Para isso, vamos precisar filtrar. A forma de fitlrar o dataframe é uma simples comparação
- sabemos que o id da loja é o 306

````python
vendas_lojacontoso = vendas_df[vendas_df['ID Loja'] == 306]
display(vendas_lojacontoso)
````

````python
qtde_vendida = vendas_lojacontoso['Quantidade Vendida'].sum()
qtde_devolvida = vendas_lojacontoso['Quantidade Devolvida'].sum()

print(f'{qtde_devolvida/qtde_vendida:.2%}')
````

## Vamos fazer em 2 passos para entender o que esta acontecendo

## Desafio: se eu quisesse criar uma tabela apenas com as vendas da Loja Contoso Europe Online e que não tiveram nenhuma devolução. Quero criar essa tabela e saber quantas vendas são.
- Repare que nesse caso são 2 condições

````python
# tudo junto
loja306_semdev = vendas_df[(vendas_df['ID Loja'] == 306) & (vendas_df['Quantidade Devolvida'] == 0)]
display(loja306_semdev)
````

````python
#separado
loja306 = vendas_df['ID Loja'] == 306
sem_dev = vendas_df['Quantidade Devolvida'] == 0
dataframe = vendas_df[loja306 & sem_dev]
display(dataframe)
````

# Adicionando Colunas, Modificando colunas e valores

````python
import pandas as pd
# importando arquivos
vendas_df = pd.read_csv(r'arquivos/Contoso - Vendas - 2017.csv', sep=';')
produtos_df = pd.read_csv(r'arquivos/Contoso - Cadastro Produtos.csv', sep=';')
lojas_df = pd.read_csv(r'arquivos/Contoso - Lojas.csv', sep=';')
clientes_df = pd.read_csv(r'arquivos/Contoso - Clientes.csv', sep=';')

# limpando apenas as colunas que queremos
clientes_df = clientes_df[['ID Cliente', 'E-mail']]
produtos_df = produtos_df[['ID Produto', 'Nome da Marca']]
lojas_df = lojas_df[['ID Loja', 'Nome da Loja']]

# mesclando e renomeando os dataframes
vendas_df = vendas_df.merge(produtos_df, on='ID Produto')
vendas_df = vendas_df.merge(lojas_df, on='ID Loja')
vendas_df = vendas_df.merge(clientes_df, on='ID Cliente')
vendas_df = vendas_df.rename(columns={'E-mail': 'E-mail Cliente'})
display(vendas_df)
````

## Agora, se quisermos acrescentar uma coluna com o mês, o dia e o ano de cada venda(e não só a data completa)

````python
vendas_df['Data da Venda'] = pd.to_datetime(vendas_df['Data da Venda'], format='%d/%m/%Y')
vendas_df['Ano da Venda'] = vendas_df['Data da Venda'].dt.year
vendas_df['Mês da Venda'] = vendas_df['Data da Venda'].dt.month
vendas_df['Dia da Venda'] = vendas_df['Data da Venda'].dt.day
display(vendas_df)
vendas_df.info()
````
